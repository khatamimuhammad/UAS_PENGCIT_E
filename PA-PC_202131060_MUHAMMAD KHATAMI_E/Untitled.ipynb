# DETEKSI BENTUK GAMBAR DENGAN CONTOUR 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Membaca gambar
image = cv2.imread('image.jpg')
```

```python
# Mengubah gambar menjadi grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Menerapkan thresholding
ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
```

```python
# Mendapatkan kontur
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
# Menampilkan gambar asli dengan kontur
image_contour = image.copy()
cv2.drawContours(image_contour, contours, -1, (0, 255, 0), 2)
```

```python
# Menampilkan gambar
plt.figure(figsize=(8, 6))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Gambar Asli')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(image_contour, cv2.COLOR_BGR2RGB))
plt.title('Deteksi Contour')

plt.tight_layout()
plt.show()
```
